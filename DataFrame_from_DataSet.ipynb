```python
import pandas as pd
pd.set_option("max_colwidth", None)
import os
```

```python
folder_path = '/content/drive/MyDrive/Projet_Plante/New Plant Diseases Dataset(Augmented)/'
```

```python
#Extraction
data = []
t = 0
for path, dirs, files in os.walk(folder_path):
  for filename in files:
    if filename[-3:] == 'JPG': 
      data_temp = []
      if path.split('/')[7] == "train" :
        data_temp.append("train")
      else :
        data_temp.append("valid")                
      data_temp.append(path  +'/' +  filename)
      data_temp.append(path.split('/')[-1].split('___')[0])
      data_temp.append(path.split('/')[-1].split('___')[1])
      if path.split('/')[-1].split('___')[1] == 'healthy' :
        data_temp.append(1)
      else :
        data_temp.append(0)
      data.append(data_temp)
      # t+=1
      # if t > 10 :
      #   print(data)
      #   break
     
      
    
    
```

```python

df = pd.DataFrame(data,columns=['train/valid','Directory','type de plante','maladie','healthy'])
df['class'] =  df['type de plante'] +'/' + df['maladie']
df['class'] = df['class'].astype('category')
df['class'] = df['class'].cat.codes
df['pre_process'] = df['Directory'].apply(lambda x : False if x[-7] in str(list(range(0,10))) else True)
df.head()
```

```python
df.to_csv('/content/drive/MyDrive/Projet_Plante/Metadata_Keras.csv')
```

```python
classe_plante = df
# classe_plante.set_index(classe_plante['class'],inplace=True) 
classe_plante = classe_plante.drop(['train/valid','Directory','pre_process'],1)

liste = []

def f(x):
  ret = False
  global liste
  if x not in liste :
    ret = True 
    liste.append(x)
  return ret

classe_plante['temp'] = classe_plante['class'].apply(lambda x: f(x))
```

```python
classe_plante = classe_plante[classe_plante['temp']==True]
classe_plante.drop('temp',1,inplace=True)
classe_plante.sort_values('class',inplace=True)
classe_plante = classe_plante.set_index('class')
classe_plante
```
